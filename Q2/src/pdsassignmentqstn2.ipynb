```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('rawdata/StudentsPerformance.csv')
data.head()
```

```python
data.info()
```

# Visualizations

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.hist(data['gender'])
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.subplot(2,2,2)
plt.hist(data['race/ethnicity'])
plt.title('Distribution of race/ethnicity')
plt.xlabel('race/ethnicity')
plt.subplot(2,2,3)
plt.hist(data['parental level of education'])
plt.title('Distribution of Parent Background')
plt.xlabel('parental level of education')
plt.subplot(2,2,4)
plt.hist(data['lunch'])
plt.title('Distribution of lunch')
plt.xlabel('lunch')
plt.savefig('results/visualization_1.png')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.hist(data['math score'])
plt.title('Distribution of Math scores')
plt.xlabel('Math Score')
plt.subplot(1,3,2)
plt.hist(data['reading score'])
plt.title('Distribution of Reading score')
plt.xlabel('Reading Score')
plt.subplot(1,3,3)
plt.hist(data['writing score'])
plt.title('Distribution of writing score')
plt.xlabel('Writing Score')
plt.savefig('results/visualization_2.png')
plt.show()
```

```python
plt.figure(figsize=(15,20))
plt.subplot(3,1,1)
plt.scatter(data['math score'], data['reading score'])
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.title('Math Score vs Reading Score')
plt.subplot(3,1,2)
plt.scatter(data['reading score'], data['writing score'])
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.title('Reading Score vs Writing Score')
plt.subplot(3,1,3)
plt.scatter(data['math score'], data['writing score'])
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.title('Math Score vs Writing Score')
plt.savefig('results/visualization_3.png')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.boxplot(data['math score'])
plt.xlabel('Maths')
plt.subplot(1,3,2)
plt.boxplot(data['reading score'])
plt.xlabel('Reading')
plt.subplot(1,3,3)
plt.boxplot(data['writing score'])
plt.xlabel('Writing')
plt.savefig('results/visualization_4.png')
plt.show()
```

```python
prep_course_count = data['test preparation course'].value_counts()
plt.bar(prep_course_count.index, prep_course_count.values)
plt.xlabel('Test Preparation Course')
plt.ylabel('Number of Students')
plt.title('Number of Students Who Completed Test Preparation Course')
plt.savefig('results/visualization_5.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='math score', data=data)
plt.xlabel('Parental Level of Education')
plt.ylabel('Math Score')
plt.title('Box plot of Math Scores by Parental Level of Education')
plt.xticks(rotation=45)
plt.savefig('results/visualization_6.png')
plt.show()
```

```python
plt.figure(figsize=(15,20))
plt.scatter(data['math score'], data['reading score'])
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.title('Math Score vs Reading Score')
plt.show()
```

```python
counts = data['race/ethnicity'].value_counts()

plt.bar(counts.index, counts.values)
plt.xlabel('Race/Ethnicity')
plt.ylabel('Number of Students')
plt.title('Number of Students in each Race/Ethnicity Category')
plt.show()
```

```python
plt.hist(data['writing score'], bins=10)
plt.xlabel('Writing Score')
plt.ylabel('Frequency')
plt.title('Distribution of Writing Scores')
plt.show()
```

```python
plt.boxplot([data[data['gender'] == 'female']['math score'], data[data['gender'] == 'male']['math score']])
plt.xticks([1, 2], ['Female', 'Male'])
plt.ylabel('Math Score')
plt.title('Distribution of Math Scores by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='reading score', data=data)
plt.xlabel('Parental Level of Education')
plt.ylabel('Reading Score')
plt.title('Box plot of Reading Scores by Parental Level of Education')
plt.show()
```

```python
sns.pairplot(data, hue='gender')
plt.show()
```

```python
sns.violinplot(x='parental level of education', y='reading score', data=data)
plt.show()
```
